```python
import os

# Count the number of donors by sex
donor_counts = sample_df['sex'].value_counts()

# Define custom colors for the bars
bar_colors = ['grey']

# Set the plot size
fig, ax = plt.subplots(figsize=(2, 3))

# Create a bar plot with custom colors
ax.bar(donor_counts.index, donor_counts.values, color=bar_colors)

# Customize the plot
ax.set_xlabel('Sex')
ax.set_ylabel('Number of Donors')
ax.set_title('Number of Donors by Sex')

# Save the plot as an image file (e.g., PDF)
plt.savefig(os.path.join(fig_dir, 'donor_counts_by_sex.pdf'))

# Show the plot
plt.show()
```
